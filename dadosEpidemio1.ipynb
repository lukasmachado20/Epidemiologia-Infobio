```python
import numpy as np
import pandas as pd
```

### Pegando conjunto de dados e colocando em formato de pandas 

```python
dataFrame1 = pd.read_excel("Arquivos/EPI_INFOBIO_CIRURGIA 2014 - 2021a.xlsx")
columnsDF1 = np.array(dataFrame1.columns)
dataFrame1
```

```python
dataFrame2 = pd.read_excel("Arquivos/EPI_INFOBIO_CIRURGIA 2014 - 2021b (2).xlsx")
columnsDF2 = np.array(dataFrame2.columns)
dataFrame2
```

### Concatenando os dois dataframes

```python
# Une os dois dataframes e ignora colunas que possam não existir em um deles, no nosso caso o dataframe2 possui uma coluna a mais
dataFrameCompleto = pd.concat([dataFrame1, dataFrame2], ignore_index=True, sort=False)
# Estou testando para verificar se a coluna tem alguma valor, pois esta coluna só existia no dataframe2
colunaNaN = np.array(dataFrameCompleto.iloc[:,25])
np.where(colunaNaN != 0)
```

### Dataframe completo depois de concatenar os dois
* Agora pode tratar os dados inconsistêntes como o NaN, Datas, e etc..

```python
# Criando coluna de data_alta
dataFrameCompleto['data_alta'] = dataFrameCompleto['data_alta_obito']
dataFrameCompleto.loc[dataFrameCompleto['obito_hosp'] == 1,['data_alta']] = '-'
dataFrameCompleto.loc[dataFrameCompleto['obito_hosp'] == 0,['data_alta_obito']] = '-'
```

```python
dataFrameCompleto['data_alta'] = pd.to_datetime(dataFrameCompleto['data_alta'], errors='coerce')
dataFrameCompleto['data_alta_obito'] = pd.to_datetime(dataFrameCompleto['data_alta_obito'], errors='coerce')
dataFrameCompleto
```

```python
# Colocando data no formato brasileiro, azar
dataFrameCompleto.loc[:,'data_alta'] = dataFrameCompleto.loc[:,'data_alta'].dt.strftime('%d/%m/%Y')
```

```python
dataFrameCompleto.loc[:,'data_inter'] = dataFrameCompleto.loc[:,'data_inter'].dt.strftime('%d/%m/%Y')
dataFrameCompleto.loc[:,'data_cirurgia'] = dataFrameCompleto.loc[:,'data_cirurgia'].dt.strftime('%d/%m/%Y')
dataFrameCompleto.loc[:,'data_alta_obito'] = dataFrameCompleto.loc[:,'data_alta_obito'].dt.strftime('%d/%m/%Y')
```

```python
dataFrameCompleto.rename(columns={'data_alta_obito':'data_obito'}, inplace=True)
```

```python
listColumns = list(dataFrameCompleto.columns)
listColumns.insert(4,'data_alta')
listColumns.pop(-1)
```

```python
# Reordenando colunas
dataFrameCompleto = dataFrameCompleto.loc[:,listColumns]
dataFrameCompleto
```

### Tratando NaN

```python
dataFrameCompleto.fillna('-', inplace=True)
dataFrameCompleto[dataFrameCompleto.isin(['-']).any(axis=1)]
```
